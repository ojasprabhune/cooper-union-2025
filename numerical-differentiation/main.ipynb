```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_fnc(graph_pos, x, y, special_y, dx, num_dx, z_lim, x_label, y_label, title, ax):
    ax[graph_pos[0], graph_pos[1]].plot(x, y)
    ax[graph_pos[0], graph_pos[1]].plot(x, special_y / dx ** num_dx)
    ax[graph_pos[0], graph_pos[1]].set_ylim(-z_lim, z_lim)
    ax[graph_pos[0], graph_pos[1]].set_xlabel(x_label)
    ax[graph_pos[0], graph_pos[1]].set_ylabel(y_label)
    ax[graph_pos[0], graph_pos[1]].set_title(title)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.subplots(2, 2)

size = 1
z_lim = 1.1
resolution = 100

x = np.linspace(-size * np.pi, size * np.pi, resolution)
dx = x[1] - x[0]

x1 = np.linspace(-size * np.pi, size * np.pi, resolution + 1)
x2 = np.linspace(-size * np.pi, size * np.pi, resolution + 2)
x3 = np.linspace(-size * np.pi, size * np.pi, resolution + 3)
x4 = np.linspace(-size * np.pi, size * np.pi, resolution + 4)

y1 = np.sin(x1)
y2 = np.sin(x2)
y3 = np.sin(x3)
y4 = np.sin(x4)

ny1 = np.sin(x)
ny2 = np.cos(x)
ny3 = -np.sin(x)
ny4 = -np.cos(x)

cos_y =         np.diff(y1)
neg_sin_y =     np.diff(np.diff(y2))
neg_cos_y =     np.diff(np.diff(np.diff(y3)))
insane_sin_y =  np.diff(np.diff(np.diff(np.diff(y4))))

plot_fnc([0, 0], x, ny1, insane_sin_y, dx, 4, z_lim,  "x", "y", "sin x and 4th derivative of y", ax)
plot_fnc([0, 1], x, ny2, cos_y, dx, 1, z_lim,         "x", "y", "cos x and 1st derivative of y", ax)
plot_fnc([1, 0], x, ny3, neg_sin_y, dx, 2, z_lim,     "x", "y", "-sin x and 2nd derivative of y", ax)
plot_fnc([1, 1], x, ny4, neg_cos_y, dx, 3, z_lim,     "x", "y", "-cos x and 3rd derivative of y", ax)

plt.tight_layout() 
plt.show()
```
